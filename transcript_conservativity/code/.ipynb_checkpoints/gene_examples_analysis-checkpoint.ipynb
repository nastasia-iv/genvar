# Analysis of selected genes

We selected four genes based on previous analysis: HLA-DRA, CFLAR, TPT1, and RPS10

- [**HLA-DRA**](#gene1) a key gene in the MHC class II region, encodes a major histocompatibility complex (MHC) class II alpha chain. This gene is essential for antigen presentation to T cells, playing a pivotal role in the adaptive immune response.
- [**CFLAR**](#gene2) also known as CASP8 and FADD-like apoptosis regulator, is involved in regulating apoptosis and immune cell function. It serves as a critical modulator of cell survival pathways, influencing the delicate balance between cell death and proliferation.
- [**TPT1**](#gene3) or translationally controlled tumor protein, participates in multiple cellular processes, including cell growth, differentiation, and apoptosis. This gene is implicated in cancer development and progression, highlighting its significance in tumorigenesis.
- [**RPS10**](#gene4) a ribosomal protein S10, is a component of the ribosome essential for protein synthesis. It plays a fundamental role in translating genetic information into functional proteins, demonstrating its importance in maintaining cellular homeostasis.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import statistics
```

```python
gene_data_HLADRA = pd.read_csv('/home/bioinf/gnomad-variants/Grishchenko/processed_data/tables/ac_expression_tables/HLADRA_analysis.tsv', 
                         sep='\t')
gene_data_CFLAR = pd.read_csv('/home/bioinf/gnomad-variants/Grishchenko/processed_data/tables/ac_expression_tables/CFLAR_analysis.tsv', 
                         sep='\t')
gene_data_TPT1 = pd.read_csv('/home/bioinf/gnomad-variants/Grishchenko/processed_data/tables/ac_expression_tables/TPT1_analysis.tsv', 
                         sep='\t')
gene_data_RPS10 = pd.read_csv('/home/bioinf/gnomad-variants/Grishchenko/processed_data/tables/ac_expression_tables/RPS10_analysis.tsv', 
                         sep='\t')
```

## Analysis of LOEUF or Exon number per transcript versus AC/Variant for HLA-DRA, CFLAR, TPT1, and RPS10 genes

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

#     4  (df1, df2, df3, df4)
dataframes = [gene_data_HLADRA, gene_data_CFLAR, gene_data_TPT1, gene_data_RPS10]
titles = ['HLA-DRA', 'CFLAR', 'TPT1', 'RPS10']

for i, df in enumerate(dataframes):
    ax = axs[i//2, i%2]
    ax.scatter(np.log1p(df['AC/Variant']), df['lof_transcript'], color='seagreen', s = 25)
    ax.set_title(titles[i], fontsize=16, fontweight='bold')
    ax.set_xlabel('AC/Variant')
    ax.set_ylabel('LOEUF per transcript')

plt.tight_layout()
plt.savefig('../data/LOEUF_AC_plot.png', bbox_inches='tight')
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

#     4  (df1, df2, df3, df4)
dataframes = [gene_data_HLADRA, gene_data_CFLAR, gene_data_TPT1, gene_data_RPS10]
titles = ['HLA-DRA', 'CFLAR', 'TPT1', 'RPS10']

for i, df in enumerate(dataframes):
    ax = axs[i//2, i%2]
    ax.scatter(np.log1p(df['AC/Variant']), df['Exon_number'], color='mediumvioletred', s = 15)
    ax.set_title(titles[i], fontsize=16, fontweight='bold')
    ax.set_xlabel('AC/Variant')
    ax.set_ylabel('Exons in transcript')

plt.tight_layout()
plt.savefig('../data/Exon_AC_plot.png', bbox_inches='tight')
plt.show()
```
